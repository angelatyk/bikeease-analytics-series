# Task 2: Data Processing and Statistical Analysis

## Perform transformations

```python
# Multiply Temperature by 10 for standardization
df['Temperature(C) Std'] = ((df['Temperature(C)'] * 10) - (df['Temperature(C)'] * 10).mean()) / (df['Temperature(C)'] * 10).std()
```

```python
# Scale Visibility to a range between 0 and 1 using MinMax scaling
df['Visibility (10m)'] = (df['Visibility (10m)'] - df['Visibility (10m)'].min()) / (df['Visibility (10m)'].max() - df['Visibility (10m)'].min())
```

## Conduct basic statistical analysis

```python
# Use describe() function for key columns like Temperature, Humidity(%), Rented Bike Count
df[['Temperature(C) Std', 'Humidity(%)', 'Rented Bike Count']].describe()
```

Compare the results with raw dataset statistics

Based on describe(), it is visible Temperature has been transformed using standarization where as Humidity and rented Bike Count remains unchanged.

## Perform pivoting operations on the dataset based on categorical columns

```python
# Group by Seasons and calculate the average rented bike count
average_rentals_by_season = df.groupby('Seasons')['Rented Bike Count'].mean()

print('\nUsing Groupby:')
print(average_rentals_by_season)

season_pivot = df.pivot_table(index='Seasons', values='Rented Bike Count', aggfunc='mean')

print(season_pivot)

# it is more popular to rent bikes during hotter or warmer seasons such as Summer and Autumn.
# it is least popular to rent bikes during colder season like Winter.
```

```python
# Group by Holiday and calculate the average rented bike count
holiday_trend = df.groupby('Holiday')['Rented Bike Count'].mean()

print('\nUsing Groupby:')
print(holiday_trend)

holiday_pivot = df.pivot_table(index='Holiday', values='Rented Bike Count', aggfunc='mean')

print('\nUsing Pivot Table:')
print(holiday_pivot)

# it is more popular to rent bikes when it is not a Holiday (surprisingly!)
```

```python
# Group by Functioning Day and calculate the average rented bike count
functioning_day_trend = df.groupby('Functioning Day')['Rented Bike Count'].mean()

print('\nUsing Groupby:')
print(functioning_day_trend)

functioning_day_pivot = df.pivot_table(index='Functioning Day', values='Rented Bike Count', aggfunc='mean')

print('\nUsing Pivot Table:')
print(functioning_day_pivot)

# no bikes are rented on non functioning days
```

## Create distribution tables

### Temperature and Rented Bike Count distribution by Hour

```python
# group by Hour and calculate averages
temp_distribution = df.groupby('Hour')['Temperature(C)'].mean().reset_index(name='Avg Temperature (C)')
bike_count_distribution = df.groupby('Hour')['Rented Bike Count'].mean().reset_index(name='Avg Rented Bike Count')

# merge into one table
distribution_table = pd.merge(temp_distribution, bike_count_distribution, on='Hour')

print(distribution_table)
```

I initially thought distribution table meant graphing it...I later realized it meant a table to summarize the data distributed...but the graph looks so pretty so I kept it here anyways.

```python
sns.set_theme(style='whitegrid')

# create temperature bins, this groups the temperatures into ranges
df['Temp Bin'] = pd.cut(df['Temperature(C)'], bins=3)

# group by Hour and Temp Bin, then calculate mean
grouped = df.groupby(['Hour', 'Temp Bin'], observed=True)['Rented Bike Count'].mean().reset_index()

plt.figure(figsize=(12, 8))

sns.barplot(
    data=grouped,
    x='Hour',
    y='Rented Bike Count',
    hue='Temp Bin',
    palette='viridis'
)

# add title and labels
plt.title('Average Rented Bike Count by Hour and Temperature Range')
plt.xlabel('Hour of Day')
plt.ylabel('Average Rented Bike Count')
plt.legend(title='Temperature Range (°C)', loc='upper left')

plt.tight_layout()
plt.show()
```

# Task 4: Data Visualization

## Import visualization libraries (matplotlib, seaborn)

```python
# was already done at the top and normally would not do again down here...but for completion
import matplotlib.pyplot as plt
import seaborn as sns
```

## Select appropriate visualization techniques for the data

### Bar plot for average rentals by Seasons

```python
sns.set_theme(style='whitegrid')

plt.figure(figsize=(8, 5))

sns.barplot(
    data=df,
    x='Seasons',
    y='Rented Bike Count',
    hue='Seasons',
    palette='Set2'
)

# add title and labels
plt.title('Average Rented Bike Count by Season')
plt.ylabel('Average Rented Bike Count')
plt.xlabel('Season')

plt.tight_layout()
plt.show()

plt.savefig('rental_vs_seasons.png', dpi=300, bbox_inches='tight')
plt.close()
```

### Line plot showing hourly rentals throughout the day

```python
sns.set_theme(style='whitegrid')

# group by hour and calculate average rented bike count
hourly_rentals = df.groupby('Hour')['Rented Bike Count'].mean().reset_index()

plt.figure(figsize=(12, 6))

sns.lineplot(
    data=hourly_rentals,
    x='Hour',
    y='Rented Bike Count',
    color='red'
)

# add title and labels
plt.title('Average Hourly Bike Rentals Throughout the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Rented Bike Count')
plt.xticks(range(0, 24))
plt.grid(True)

plt.tight_layout()
plt.show()

plt.savefig('average_hourly_rentals.png', dpi=300, bbox_inches='tight')
plt.close()
```

### Heatmap showing correlation among numerical variables

```python
sns.set_theme(style='whitegrid')

# select only numerical columns
numerical_df = df.select_dtypes(include='number')

# compute correlation matrix
corr_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    square=True,
    linewidths=0.5,
    cbar_kws={'shrink': 0.8}
)

# add title
plt.title('Correlation Heatmap of Numerical Variables')

plt.tight_layout()
plt.show()

plt.savefig('numerical_variables_heatmap.png', dpi=300, bbox_inches='tight')
plt.close()
```

### Box plot to identify outliers in Temperature and Rented Bike Count

```python
sns.set_theme(style='whitegrid')

# create temperature bins, this groups the temperatures into ranges
df['Temp Bin'] = pd.cut(df['Temperature(C)'], bins=5)

plt.figure(figsize=(10, 5))

sns.boxplot(
    data=df,
    x='Temp Bin',
    y='Rented Bike Count',
    hue='Temp Bin',
    palette='Set2',
    legend=False
)

# add title and labels
plt.title('Rented Bike Count Distribution Across Temperature Ranges')
plt.xlabel('Temperature Range (°C)')
plt.ylabel('Rented Bike Count')

plt.tight_layout()
plt.show()

plt.savefig('rental_vs_temperature.png', dpi=300, bbox_inches='tight')
plt.close()
```

## Record observations and insights from visualizations

Based on the average rentals by season plot, warmer seasons such as summer and autumn are more favorable for biking.

According to the line plot showing hourly rentals throughout the day, there are more bike rentals in the evening than in the morning. The lowest rental activity occurs late at night.

From the heatmap showing correlations among numerical variables, warmer and clearer weather conditions tend to increase bike rentals, while higher humidity and precipitation are associated with lower rental counts.

Based on the box plot of temperature vs. rented bike count, people rent more bikes during warmer weather, with higher demand observed at higher temperatures.
